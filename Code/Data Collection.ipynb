```python
data = '../Data/'
```

```python
# concat all csv files from season 1989-2022
csv_files = [f for f in os.listdir(data) if f.endswith('.csv')]
dfs = []
for file in csv_files:
    file_path = os.path.join(data, file)
    df = pd.read_csv(file_path, header = None)
    dfs.append(df)

combo_df = pd.concat(dfs, ignore_index=True)
```

```python
combo_df
```

```python
combo_df.columns = column_names
```

```python
combo_df
```

```python
#order df by date and reset index to reorder the games by date
combo_df = combo_df.sort_values(by='date').reset_index(drop=True)
```

```python
combo_df
```

```python
# create new column for run differential
combo_df['run_diff'] = combo_df['h_runs'] - combo_df['v_runs']
```

```python
combo_df
```

```python
combo_df.isna().sum()
```

```python
# taking out columns with too many missing values
combo_df = combo_df.drop(columns = ['make_up', 'forfeit', 'protest', 'misc', 'lf_ump_id', 'rf_ump_id', 'lf_ump_name', 'rf_ump_name'])
```

```python
combo_df
```

```python
#replace relevant missing columns with appropriate value. ie 'None' or 0
combo_df['attndnce'].fillna(0, inplace=True)
combo_df['2b_ump_id'].fillna('None', inplace=True)
combo_df['sve_pitch_id'].fillna('None', inplace=True)
combo_df['gwrbi_id'].fillna('None', inplace=True)
combo_df['3b_ump_id'].fillna('None', inplace=True)
combo_df['win_pitcher_id'].fillna('None', inplace=True)           
combo_df['win_pitcher_name'].fillna('None', inplace=True)
combo_df['lose_pitch_name'].fillna('None', inplace=True)
combo_df['lose_pitch_id'].fillna('None', inplace=True)
```

```python
# save df as a checkpoint
combo_df.to_csv('../Data/NoMissingGames.csv', index=False)
```
