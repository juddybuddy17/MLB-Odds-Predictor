```python
team_abb = 'BAL', 'CLE', 'KCA', 'OAK', 'CIN', 'NYN', 'SDN', 'TEX', 'CAL','MIN', 'MON', 'HOU', 'CHN', 'NYA', 'DET', 'ATL', 'PHI', 'SFN','SEA', 'BOS', 'MIL', 'PIT', 'LAN', 'CHA', 'SLN', 'TOR', 'FLO','COL', 'ANA', 'TBA', 'ARI', 'WAS', 'MIA'
```

```python
teams = teams[teams['TEAM'].isin(team_abb)]
teams['team'] = teams['TEAM']
```

```python
teams
```

```python
teams['team'] = teams['TEAM'].replace(team_names)
```

```python
teams = teams['team'].reset_index()
```
