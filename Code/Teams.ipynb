# Home and Away Cleaning Data

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 90)
pd.set_option('display.max_rows',500)
```

```python
games = pd.read_csv('../Data/Created/MLBgames.csv')
```

```python
games['home_team'].unique()
```

```python
# I am a SF Giants fan so let's take a look at them
sf_giants = games.loc[((games['visitor_team'] == 'SFN') | (games['home_team'] == 'SFN')), :]
sf_giants
```

```python
sf_wins = games[games['W'] == 'SFN']['W'].count()
sf_loss = games[games['L'] == 'SFN']['L'].count()
print(sf_wins, sf_loss)
# just for my enjoyment, in the time span of this df the giants record is 2775 and 2564
```

```python
# Make a data frame of giants home team
def sf_home_team(df, team_code='SFN'):
    return df[df['home_team'] == team_code]

sf_home_games = sf_home_team(sf_giants)
sf_home_games
```

```python
def consolidate_sf(df):
    
    consol_df = df.groupby('season_year').first()
    consol_df = consol_df[['home_team', 'h_roll_ba','h_roll_obp','h_roll_tbs','h_roll_slg','h_roll_ops','h_roll_k_rat','h_roll_rbis','h_roll_bb_rat','h_roll_sb',
                           'h_roll_hr','h_roll_lob']]
    return consol_df
```

```python
sf_consol = consolidate_sf(sf_home_games)
sf_consol
```

```python
# now do it for each team
games
```

```python
home_games = games.groupby(['home_team', 'season_year']).agg({
    'h_roll_ba': 'mean',
    'h_roll_obp': 'mean',
    'h_roll_tbs': 'sum',
    'h_roll_slg': 'mean',
    'h_roll_ops': 'mean',
    'h_roll_k_rat': 'mean',
    'h_roll_rbis': 'sum',
    'h_roll_bb_rat': 'mean',
    'h_roll_sb': 'sum',
    'h_roll_hr': 'sum',
    'h_roll_lob': 'mean',
    'h_roll_runs': 'sum'
}).reset_index()
home_games
```

```python
away_games = games.groupby(['visitor_team', 'season_year']).agg({
    'v_roll_ba': 'mean',
    'v_roll_obp': 'mean',
    'v_roll_tbs': 'sum',
    'v_roll_slg': 'mean',
    'v_roll_ops': 'mean',
    'v_roll_k_rat': 'mean',
    'v_roll_rbis': 'sum',
    'v_roll_bb_rat': 'mean',
    'v_roll_sb': 'sum',
    'v_roll_hr': 'sum',
    'v_roll_lob': 'mean',
    'v_roll_runs': 'sum'
}).reset_index()
away_games
```

```python
away_games.drop(columns = ['season_year'], inplace=True)
```

```python
teams = pd.concat([home_games, away_games], axis=1)
teams
```
